## Sampling Distribution of the mean


```python
exp_dist = np.random.exponential(scale=1., size=10000)
pylab.hist(exp_dist, bins=50)
print()
```

Polulation은 exponential distribution을 따라갈지라도, sample mean $ \bar{x} $ 는 정규분포를 따릅니다.<br>
이때 poluation이 동일하게 normally distributed 라면 sample의 크기와 상관없이 정규분포를 따르며, <br>
아닐경우 sample의 갯수 n의 크기가 30개 이상일때 대략 sample mean은 population distribution이 뭐든간에 정규분포를 따릅니다.

```python
fig, plots = plt.subplots(3, 3)
fig.set_figheight(7)
fig.set_figwidth(10)
fig.tight_layout()
plots = plots.reshape(-1)

for i, p in enumerate(plots):
    size = 1 if i == 0 else i * 2
    sample_means = [np.mean(np.random.choice(exp_dist, size=size, replace=False)) for _ in range(2000)]
    p.hist(sample_means, bins=40)
    p.set_title(f'size={size}')
```
