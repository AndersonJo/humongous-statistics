# Gaussian Function

### 1-dimension 

Gaussian function은 **[normal distribution](http://mathworld.wolfram.com/NormalDistribution.html)**, Bell Curve 또는 Probability Density라고도 불립니다. 

### $$ f(x) =  e^{-\frac{(x- \mu)^2}{2\sigma^2}} $$

```python
%pylab inline
import numpy as np
from scipy import signal
from mpl_toolkits.mplot3d import Axes3D
```

```python
def gaussian(x, std=None):
    mean = np.mean(x)
    if std is None:
        std = np.std(x)
    return np.exp(-(x-mean)**2/(2 * std**2))

plot(gaussian(np.linspace(-3, 3, 120), std=0.1), color='green')
plot(gaussian(np.linspace(-3, 3, 120), std=0.5), color='blue')
plot(gaussian(np.linspace(-3, 3, 120), std=1), color='red')

print signal.gaussian(6, std=7) # Scipy Gaussian
print gaussian(np.arange(6), std=7) # My Gaussian
```

### 2-dimensional

### $$ f(x, y) = e^{-  \frac{(x-\mu_x)^2 + (y - \mu_y)^2 }{2 \sigma^2}  } $$

```python
def gaussian2d(x, y, std=None):
    if std is None:
        std = x.std()
    xmean = x.mean()
    ymean = y.mean()
    return np.exp(-((x-xmean)**2 + (y-ymean)**2)/(2*std**2) )
    
X = np.linspace(-3, 3, 120)
Y = np.linspace(-3, 3, 120)

plot(gaussian2d(X, Y, std=0.1), color='green')
plot(gaussian2d(X, Y, std=0.5), color='blue')
plot(gaussian2d(X, Y, std=1), color='red')
```
